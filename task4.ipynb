```python
import json
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

class QuizApp:
    def __init__(self, questions_data):
        self.questions_data = questions_data
        self.current_question = None
        self.selected_answer = None

        # Create widgets
        self.question_number_label = widgets.Label(value="")
        self.question_text_label = widgets.Label(value="")
        self.answer_buttons = [widgets.Button(description="", layout=widgets.Layout(width='auto')) for _ in range(4)]
        self.submit_button = widgets.Button(description="Submit", layout=widgets.Layout(width='auto'))
        self.output = widgets.Output()

        # Assign callback function for button clicks
        for i, button in enumerate(self.answer_buttons):
            button.on_click(lambda x, i=i: self.select_answer(i + 1))

        self.submit_button.on_click(self.check_answer)

        # Display widgets
        display(self.question_number_label)
        display(self.question_text_label)
        display(widgets.VBox(self.answer_buttons))
        display(self.submit_button)
        display(self.output)

        # Start the quiz
        self.load_next_question()

    def load_next_question(self):
        self.current_question = random.choice(self.questions_data)

        # Display question information
        self.question_number_label.value = f"Question {self.current_question['question_number']}"
        self.question_text_label.value = self.current_question['question_text']

        for i, button in enumerate(self.answer_buttons):
            button.description = self.current_question['answers'][i]

    def select_answer(self, option):
        self.selected_answer = option

    def check_answer(self, button):
        # Check if an answer is selected
        if self.selected_answer is None:
            with self.output:
                clear_output(wait=True)
                print("Please select an answer.")
            return

        # Check if the selected answer is correct
        selected_option = self.selected_answer
        correct_option = self.current_question['correct_answer']
        correct_option=ord(correct_option) - ord('A') + 1


        with self.output:
            clear_output(wait=True)
            if selected_option == correct_option:
                print("Congratulations! Your answer is correct.")
            else:
                print(f"Sorry, your answer is incorrect. The correct answer is option {correct_option}")

        # Load the next question
        self.load_next_question()

# Load questions from JSON
with open('question.json', 'r') as file:
    questions_data = json.load(file)["questions"]

# Create QuizApp instance
quiz_app = QuizApp(questions_data)
```
