```python
from google.colab import drive
import ipywidgets as widgets
from ipywidgets import HBox, VBox
import json
from IPython.display import display
updated_json_string=''
student_data=''
def extract_new_info():
  student_data=''
  with open('json.json', 'r') as json_file:
    try:
        student_data = json_file.read()
    except json.JSONDecodeError:
        student_data = {}


def update_json_object(json_string, name_to_update, field_to_update, string_to_update):
  try:
      data = json.loads(json_string)

      for obj in data:
          if obj.get('firstname') == name_to_update:
              obj[field_to_update] = string_to_update
              break

      updated_json_string = json.dumps(data)
      with open('json.json', 'w') as json_file:
        json_file.write(updated_json_string)
      global student_data
      student_data=''
      student_data=updated_json_string
      return updated_json_string
  except json.JSONDecodeError:
      return "Error: Invalid JSON input"



def extract_info(json_string, target_name):
    try:
        data = json.loads(json_string)
        records_for_name = [record for record in data if record.get('firstname') == target_name]
        values_only = [list(record.values()) for record in records_for_name]
        return values_only
    except json.JSONDecodeError:
        return "Error: Invalid JSON input"


extract_new_info()

with open('json.json', 'r') as json_file:
    try:
        student_data = json_file.read()
    except json.JSONDecodeError:
        student_data = {}
data = json.loads(student_data)
first_names = [entry['firstname'] for entry in data]

global selected_name
def on_dropdown_change(change):
    selected_name = change['new']
    extract_new_info()
    result = extract_info(student_data, selected_name)
    print(result[0])
    label_right.value = f'{result[0][0]}'
    label_right_2.value = f'{result[0][1]}'
    label_right_3.value = f'{result[0][2]}'
    label_right_4.value = f'{result[0][3]}'
    label_right_5.value = f'{result[0][4]}'
    label_right_6.value = f'{result[0][5]}'




dropdown = widgets.Dropdown(
    options=first_names,
    value=None,
    description='Select a name:',
)
dropdown.style.description_width = '100px'
dropdown.observe(on_dropdown_change, names='value')

label_left = widgets.Label(value='First Name:')
label_right = widgets.Label()
label_left_2 = widgets.Label(value='Last Name:')
label_right_2 = widgets.Label()
label_left_3 = widgets.Label(value='Email:')
label_right_3 = widgets.Label()
label_left_4 = widgets.Label(value='Courses')
label_right_4 = widgets.Label()
label_left_5 = widgets.Label(value='Link:')
label_right_5 = widgets.Label()
label_left_6 = widgets.Label(value='favorite:')
label_right_6 = widgets.Label()

button1 = widgets.Button(description='Update')
updatetext1 = widgets.Textarea()

button2 = widgets.Button(description='Update')
updatetext2 = widgets.Textarea()

button3 = widgets.Button(description='Update')
updatetext3 = widgets.Textarea()

button4 = widgets.Button(description='Update')
updatetext4 = widgets.Textarea()

button5 = widgets.Button(description='Update')
updatetext5 = widgets.Textarea()

button6 = widgets.Button(description='Update')
updatetext6 = widgets.Textarea()

hbox_6 = HBox([ updatetext6,button6,label_left_6, label_right_6])
hbox_5 = HBox([ updatetext5,button5,label_left_5, label_right_5])
hbox_4 = HBox([updatetext4,button4, label_left_4, label_right_4 ])
hbox_3 = HBox([updatetext3,button3, label_left_3, label_right_3 ])
hbox_2 = HBox([updatetext2,button2, label_left_2, label_right_2])
hbox_1 = HBox([ label_left, label_right])

vbox = VBox([hbox_1, hbox_2, hbox_3, hbox_4,hbox_5,hbox_6])
display(dropdown, vbox)

def update_info(button, label_left, label_right, updatetext, field_to_update):
  selected_name = dropdown.value
  update_value = updatetext.value
  updatetext2.value=''
  updatetext3.value=''
  updatetext4.value=''
  updatetext5.value=''
  update_json_object(student_data, selected_name, field_to_update, update_value)
  student_data=''
  updatetext.value=''
  with open('json.json', 'r') as json_file:
    student_data = json_file.read()


button2.on_click(lambda btn: update_json_object(student_data,dropdown.value ,'lastname', updatetext2.value))
button3.on_click(lambda btn: update_json_object(student_data,dropdown.value,'mail', updatetext3.value))
button4.on_click(lambda btn: update_json_object(student_data,dropdown.value,'courses', updatetext4.value))
button5.on_click(lambda btn: update_json_object(student_data,dropdown.value,'link', updatetext5.value))
button6.on_click(lambda btn: update_json_object(student_data,dropdown.value,'favorite', updatetext6.value))
```
